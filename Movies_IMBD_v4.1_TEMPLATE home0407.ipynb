# 15. Фильмы с каким актером принесли самые высокие кассовые сборы в 2012 году? 

```python
answers['14'] = '15. Chris Hemsworth'
```

```python
films2002 = data.query('release_year==2012')
actorlist = films2002.cast.str.split('|', expand=True).stack().unique()
sumprofitmax = 0
for i in actorlist:
    c = films2002[films2002.cast.str.contains(i)].revenue.sum()
    if c <= sumprofitmax:
        continue
    else:
        sumprofitmax = c
        bmax = i
print(sumprofitmax, bmax)
```

# 16. Какой актер снялся в большем количестве высокобюджетных фильмов?

```python
answers['15'] = '16. Matt Damon'
```

```python
hbfilmsqty = 0  # Инициализация счётчика фильмов и переменной самогго успешного актёра
hbactor = 0
hbdata = data[data.budget > data.budget.mean()]

actorlist = hbdata.cast.str.cat(sep='|').split('|')

for actor in actorlist:  # перебор актёров в созданном списке актёров
    actorhbf = len(hbdata[hbdata.cast.          # количество высокобюджетных фильмов
                          str.contains(actor)])  # у этого актёра
    if hbfilmsqty > actorhbf:
        continue
    else:
        hbactor = actor  # лидер на момент итерации
        hbfilmsqty = actorhbf  # количество фильмов в которых он снялся

print(hbactor)
```

# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['23'] = '24. Four By Two Productions'
```

```python
studio = pd.DataFrame(data.production_companies.str.split(
    "|").tolist()).stack().unique()  # создаём список студий

# создаём признак количества символов в названии
data["simbolcount"] = data.original_title.str.len()

# инициализируем
amax = 0
company = 0

# ищем компанию у которой в среднем самые длинные (в символах) названия
for i in studio:
    a = data[data.production_companies.str.contains(i)].simbolcount.mean()
    if a > amax:
        amax = a
        company = i
    else:
        continue
display(company)
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
answers['24'] = '25. Midnight Picture Show'
```

```python
studio = pd.DataFrame(data.production_companies.str.split(
    "|").tolist()).stack().unique()  # создаём список студий

# создаём признак количества слов в названии
data["word_count"] = data['overview'].str.split().str.len()

# инициализируем
amax = 0
company = 0

# ищем самую многословную в названиях компанию, находя максимальное из средних количеств слов
for k in studio:
    a = data.groupby(data.production_companies.str.contains(k)
                     ).word_count.mean().max()
    if a > amax:
        amax = a
        company = k
    else:
        continue

display(company)
```
